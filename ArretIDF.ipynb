```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import urllib
```

```python
urllib.urlretrieve ("http://opendata.stif.info/explore/dataset/referentiel-arret-tc-idf/download/?format=csv&timezone=Europe/Berlin&use_labels_for_header=true", "referentiel-arret-tc-idf.csv")
rawstif = pd.read_csv('referentiel-arret-tc-idf.csv',sep=';')
rawstif.head(2)
```

```python
rawstif.describe()
```

```python
rawstif['ZDLr_LIBELLE_TYPE_ARRET'].unique()
```

```python
rawstif['X'], rawstif['Y'] = zip(*rawstif['ZDEr_X_Y'].apply(lambda x: x.split(',', 1)))
rawstif.head(2)
```

```python
rawstif.dtypes
```

```python
rawstif['X'] = pd.to_numeric(rawstif['X'])
rawstif['Y'] = pd.to_numeric(rawstif['Y'])
rawstif.dtypes
```

```python
rawstif.head(2)
```

```python
D=rawstif['LDA_LIBELLE_TYPE_ARRET'].tolist()
ColorsCode = {'Gare ferrée' : 'r', 'Arrêt de bus' :'b', 'Station de métro' : 'g', 'Arrêt de tram' : 'k'}
MarkerCode = {'Gare ferrée' : "sr", 'Arrêt de bus' : ".b", 'Station de métro' : "vk", 'Arrêt de tram' : "^k"}
SizeCode = {'Gare ferrée' : '2', 'Arrêt de bus' :'0.01', 'Station de métro' : '1', 'Arrêt de tram' : '1'}
M = [MarkerCode[arret] for arret in D]
S = [float(SizeCode[arret]) for arret in D]
D = [ColorsCode[arret] for arret in D]


type(M[10])
(M[3])
```

```python
plt.figure(figsize=(12,12)) 
plt.scatter(rawstif['X'], rawstif['Y'], s=S, color=D )
plt.axis('off')
plt.show()
```
